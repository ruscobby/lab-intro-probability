**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
from scipy import stats

# Parameters
n = 460          # tickets sold
p = 0.03         # probability a passenger misses the flight
seats = 450

# Minimum number of passengers who must miss the flight
k = n - seats    # 10

# Probability that at least k passengers miss the flight
probability = 1 - stats.binom.cdf(k - 1, n, p)

probability

```

Answer

The probability that Ironhack Airlines has seats for all passengers is approximately 88%.

This level of overbooking is statistically reasonable and acceptable in the aviation industry 

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

```python
#code here
from scipy import stats

# Probability of success
p = 0.3

# P(X â‰¥ 4): at least three failed attempts before first success
probability = 1 - stats.geom.cdf(3, p)

probability
```

Answer

The probability that the representative needs at least three attempts before success is 0.343 (34.3%).

This result follows directly from the memoryless property of the geometric distribution.

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
#code here
from scipy import stats

# Average visits per hour
lambda_visits = 500

# Server capacity
capacity = 550

# Probability the server is overwhelmed: P(X > 550)
probability = 1 - stats.poisson.cdf(capacity, lambda_visits)

probability
```

Answer

The probability that the website server is overwhelmed is approximately 1.2%.

This indicates overload is unlikely but not negligible under peak conditions.

```python
from scipy import stats

lambda_visits = 500
capacity = 550

# Probability of overload in one hour
p_hour = 1 - stats.poisson.cdf(capacity, lambda_visits)

# Probability of overload at least once in 24 hours
p_day = 1 - (1 - p_hour) ** 24

p_day
 
```

Interpretation

Even a small hourly overload probability compounds over 24 hours.
The daily risk is significantly higher than the 1-hour risk and should not be ignored in capacity planning.

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
from scipy import stats

# Average arrival rate (per minute)
lambda_rate = 1 / 10

# Time threshold (minutes)
t = 5

# Probability next customer arrives within 5 minutes
probability = stats.expon.cdf(t, scale=1/lambda_rate)

probability
```

Answer

The probability that the next customer arrives within 5 minutes is approximately 0.394 (39.4%).

This result reflects the memoryless nature of customer arrivals.

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
from scipy import stats

lambda_rate = 1/10
t = 15

# Probability no customer arrives in 15 minutes
probability_break = 1 - stats.expon.cdf(t, scale=1/lambda_rate)  # or stats.expon.sf(t, scale=1/lambda_rate)

probability_break
```

Answer

The probability that an employee can take a break is approximately 0.223 (22.3%).

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
#code here
from scipy import stats

# Parameters
mu = 150      # mean weight
sigma = 10    # standard deviation

# Range
lower = 140
upper = 160

# Probability the bird's weight is between 140 and 160 grams
probability = stats.norm.cdf(upper, mu, sigma) - stats.norm.cdf(lower, mu, sigma)

probability
```

Answer

The probability that a randomly selected bird weighs between 140 and 160 grams is approximately 68.3%.

This aligns with the empirical rule: about 68% of data falls within 1 standard deviation of the mean.

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
#code here
from scipy import stats

lambda_rate = 1/50
t = 30

# Probability component fails within 30 hours
probability = stats.expon.cdf(t, scale=1/lambda_rate)
probability
```

Answer

The probability that the component fails within the first 30 hours is approximately 0.451 (45.1%).
